### Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Importing dataset

```python
dt=pd.read_csv('groceries - groceries.csv',header=None)
dt.shape
```

```python
transactions=[]
for i in range(9836):
    transactions.append([str(dt.values[i,j])for j in range(33)])
```

### Training dataset with apriori algorithm

```python
from apyori import apriori
basket=apriori(transactions=transactions, min_support=0.002, min_confidence=0.2, _lift=3, min_lenghth=2)
```

### Results

#### Raw results

```python
results=list(basket)
results
```

#### Proper Display

```python
def inspect(results):
    product1 = []
    product2 = []
    supports = []
    confidence = []
    lift = []

    for result in results:
        try:
            if len(result) > 2 and len(result[2]) > 0:
                # Append Show1 if it's not a digit or is not a null value
                if not tuple(result[2][0][0])[0].isdigit() and tuple(result[2][0][0])[0]!='nan':
                    product1.append(tuple(result[2][0][0])[0])
                else:
                    continue
                
                # Append Show2 if it's not a digit or is not a null value
                if not tuple(result[2][0][1])[0].isdigit() and tuple(result[2][0][1])[0]!='nan':
                    product2.append(tuple(result[2][0][1])[0])
                else:
                    continue
                
                # Append other fields
                supports.append(result[1])  # Support
                confidence.append(result[2][0][2])  # Confidence
                lift.append(result[2][0][3])  # Lift
            else:
                # Handle case where structure is not as expected
                product1.append(None)
                product2.append(None)
                supports.append(None)
                confidence.append(None)
                lift.append(None)
        except IndexError:
            # Handle any index errors explicitly
            product1.append(None)
            product2.append(None)
            supports.append(None)
            confidence.append(None)
            lift.append(None)

    return list(zip(product1, product2, supports, confidence, lift))
DataFrame = pd.DataFrame(inspect(results), columns=['Product1', 'Product2', 'Support', 'Confidence', 'Lift'])
```
